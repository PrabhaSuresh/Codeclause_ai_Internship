```python
import cv2
import numpy as np
from ultralytics import YOLO
```

```python
model = YOLO('./runs/detect/train3/weights/best.pt')
```

```python
cap = cv2.VideoCapture(0)
```

```python
while True:
    net,frame = cap.read()
    if not net:
        break

    result = model.predict(frame,conf=0.5,imgsz=640)
    annotated_frame = result[0].plot()

    cv2.imshow("Gesture_Recognition",annotated_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
